```python
# EDA and cleaning 
df = df1
# renaming columns that are a bit confusing 
df.rename(columns={'cnt':'Total_amount_of_rentals'},inplace=True)

# dropping columns that are not needed 
df= df.drop('yr',axis=1)
df= df.drop('workingday',axis=1)

# Creating a new column in df with the seasons_names
season_numbers= {1:'Winter',2:'Spring',3:'Summer',4:'Fall'}
df['season_names']= df['season'].map(season_numbers)

# Creating a new column with weekdays_name
weekday_numbers = {0:'Sunday',1:'Monday',2:'Tuesday',3:'Wednesday',4:'Thursday',5:'Friday',6:'Saturday'}
df['weekday_names'] = df ['weekday'].map(weekday_numbers)

#Creating column with the month_name
df['dteday'] = pd.to_datetime(df['dteday'])
df['month_names']=df['dteday'].dt.strftime('%B')

# seperateing it based on years 2011 and 2012 
df['year']=df['dteday'].dt.year

#convert the hour column to morning - afternoon - evening - night 
def part_of_day(hour):
    if 0 <= hour < 7:
        return "Morning"
    elif 7 <= hour < 16:
        return "Afternoon"
    elif 16 <= hour < 19:
        return "Early Evening"
    elif 19 <= hour <= 23:
        return "Night"

df['time_of_day'] = df['hr'].apply(part_of_day)
df
```

```python
# Observing customer behavior 

# Looking at the amount of rentals for time of day, throughout years ( 2011 and 2012 )
rentals_by_year = df.groupby(['time_of_day','year'])['Total_amount_of_rentals'].agg(['sum','std','min','max'])
print(rentals_by_year)

# Looking at rentals throughout the course of seasons 
rentals_by_season = df.groupby(['season_names','year'])['Total_amount_of_rentals'].agg(['sum','std','min','max'])
print(rentals_by_season)
```

```python
# Looking at casual rentals throughout month within the years 2011 - 2012
casual_rentals_by_year = df.groupby(['year', 'month_names' ])['casual'].agg(['sum','min','max'])
print (casual_rentals_by_year)
```

```python
# Looking at casual rentals throughout the part of day within each month. 
casual_rentals_by_month = df.groupby(['time_of_day','year'])['casual'].agg(['sum','min','max'])
print( casual_rentals_by_month)
```

```python
# Looking at the registered rentals information 
registered_rentals_by_day = df.groupby(['time_of_day','year'])['registered'].agg(['sum','min','max'])
print(registered_rentals_by_day)
```

```python
# registered rides throughout the years
registered_rentals_by_year = df.groupby(['year', 'month_names' ])['registered'].agg(['sum','min','max'])
print(registered_rentals_by_year)
```
